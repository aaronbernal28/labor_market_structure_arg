```python
import pandas as pd


def get_ilostat_data(indicator_id, country_code=None):
	"""
	Descarga datos tabulares cruzados desde la API de ILOSTAT.

	:param indicator_id: El código del indicador en ILOSTAT (ej. EMP_2EMP_SEX_ECO_NB_A)
	:param country_code: Código ISO alfa-3 del país (ej. 'ARG' para Argentina).
	                     Si es None, trae todos los países.
	"""
	# Endpoint principal de la API de ILOSTAT
	base_url = "https://rplumber.ilo.org/data/indicator/"

	# Parámetros de la consulta
	params = [f"id={indicator_id}", "format=.csv"]
	if country_code:
		params.append(f"ref_area={country_code}")

	query_url = f"{base_url}?{'&'.join(params)}"
	print(f"Descargando datos desde: {query_url}\n")

	try:
		# Pandas puede leer directamente un CSV desde una URL
		df = pd.read_csv(
			query_url,
			storage_options={
				"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
			},
		)
		return df
	except Exception as e:
		print(f"Error al obtener los datos: {e}")
		return None
```

```python
# ==========================================
# EJEMPLOS DE INDICADORES EN ILOSTAT
# ==========================================
# EMP_2EMP_SEX_ECO_NB_A -> Empleo por Sexo y Actividad Económica (CIIU) - Anual
# EMP_2EMP_SEX_OCU_NB_A -> Empleo por Sexo y Ocupación (CIUO) - Anual

indicador_objetivo = "EMP_2EMP_SEX_ECO_NB_A"

# Ejecutar la descarga (Ejemplo filtrando por Argentina 'ARG')
df_empleo = get_ilostat_data(indicator_id=indicador_objetivo, country_code="ARG")

if df_empleo is not None:
	# Mostrar las primeras filas y las columnas disponibles
	print("Columnas de la tabla:", df_empleo.columns.tolist())
	print("\nMuestra de los datos:")
	print(df_empleo[["ref_area", "time", "sex", "classif1", "obs_value"]].head(10))

	# Opcional: Guardar a nivel local
	# df_empleo.to_csv("datos_empleo_ilo.csv", index=False)
```

```python
import pandas as pd


def fetch_detailed_ilostat(indicator, country="ARG"):
	"""
	Fetches detailed structural datasets from ILOSTAT.
	EMP_TEMP_SEX_OCU_NB_A -> Detailed Occupation (ISCO)
	EMP_TEMP_SEX_ECO_NB_A -> Detailed Industry (ISIC)
	"""
	# Using the time_comp=A for Annual data
	url = f"https://rplumber.ilo.org/data/indicator/?id={indicator}&ref_area={country}&time_comp=A&format=.csv"

	print(f"Requesting data: {url}")
	try:
		df = pd.read_csv(
			url,
			storage_options={
				"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
			},
		)

		# Filter only for ISCO-08 or ISIC Rev 4 classifications to drop legacy codes
		if "OCU" in indicator:
			df = df[df["classif1"].str.contains("ISCO08", na=False)]
		elif "ECO" in indicator:
			df = df[df["classif1"].str.contains("ISIC4", na=False)]

		return df
	except Exception as e:
		print(f"Error fetching data: {e}")
		return None
```

```python
# Fetch the detailed occupation data
df_detailed_occ = fetch_detailed_ilostat("EMP_TEMP_SEX_OCU_NB_A")

# To check the granularity available:
if df_detailed_occ is not None:
	# Look at the classification codes returned.
	# 4-digit codes in ISCO-08 usually look like 'OCU_ISCO08_4111'
	granularity_check = df_detailed_occ["classif1"].unique()
	print("Available classification nodes in this dataset:")
	print(granularity_check[:10])  # Print the first 10 to inspect the digit depth
```

```python
import pandas as pd


def scan_4_digit_nodes(countries):
	"""
	Scans ILOSTAT to count how many 4-digit occupational nodes
	are available for a list of countries.
	"""
	indicator = "EMP_TEMP_SEX_OCU_NB_A"
	results = {}

	for country in countries:
		url = f"https://rplumber.ilo.org/data/indicator/?id={indicator}&ref_area={country}&time_comp=A&format=.csv"
		try:
			df = pd.read_csv(
				url,
				storage_options={
					"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
				},
			)

			# Filter for ISCO-08 classifications
			df_isco = df[df["classif1"].str.contains("ISCO08", na=False)]

			# A 4-digit code looks like 'OCU_ISCO08_1234'
			# 'OCU_ISCO08_' is 11 characters. Plus 4 digits = 15 characters.
			nodes = df_isco["classif1"].unique()
			four_digit_nodes = [
				node for node in nodes if len(node) == 15 and node[-4:].isdigit()
			]

			results[country] = len(four_digit_nodes)
			print(f"{country}: Found {len(four_digit_nodes)} 4-digit nodes.")

		except Exception:
			print(f"{country}: Error or no detailed data available.")


# Let's test a mix of large economies and regions with strong statistical registries
target_countries = ["USA", "GBR", "CAN", "MEX", "DEU", "ZAF"]
scan_4_digit_nodes(target_countries)
```

```python
import pandas as pd


def scan_highest_digit_nodes():
	"""
	Scans ILOSTAT globally in a single optimized request to find the maximum
	occupational desegregation depth available (4-digit, 3-digit, or 2-digit).
	"""
	# EMP_2EMP_SEX_OCU_NB_A provides deeper classifications (Sub-major groups) than EMP_TEMP_SEX_OCU
	indicator = "EMP_2EMP_SEX_OCU_NB_A"

	# We drop ref_area to fetch all global records across all countries at once (much faster than a loop)
	url = f"https://rplumber.ilo.org/data/indicator/?id={indicator}&time_comp=A&format=.csv"

	print(f"Fetching global data for {indicator}...")
	try:
		df = pd.read_csv(
			url,
			storage_options={
				"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"
			},
			low_memory=False,
		)

		# Filter purely for ISCO-08 classifications
		if "classif1" in df.columns:
			df_isco = df[df["classif1"].str.contains("ISCO08", na=False)].copy()

			# Extract the raw code part: OCU_ISCO08_1234 -> 1234
			df_isco["isco_code"] = df_isco["classif1"].str.replace(
				"OCU_ISCO08_", "", regex=False
			)

			# Filter strictly numeric codes (dropping 'TOTAL' or 'X')
			df_digits = df_isco[df_isco["isco_code"].str.isdigit()].copy()
			df_digits["digit_level"] = df_digits["isco_code"].str.len()

			max_digits = df_digits["digit_level"].max()
			print(
				f"---\nMaximum digit depth found across all reporting countries: {max_digits}-digit\n"
			)

			if max_digits >= 4:
				c4 = df_digits[df_digits["digit_level"] >= 4]["ref_area"].unique()
				print(f"Countries with 4-digit desegregation: {list(c4)}")
			else:
				print(
					"Result: No country provides 4-digit occupational data in this dataset."
				)
				print(
					"ILOSTAT public bulk endpoints typically limit cross-tabulations to 2-digit (Sub-major groups)"
				)
				print(
					"to avoid suppressed cells from low sample sizes (NS/statistical significance limits).\n"
				)

				# Show the highest depth available instead
				highest_depth_countries = df_digits[
					df_digits["digit_level"] == max_digits
				]["ref_area"].unique()
				print(
					f"List of countries falling back strictly to {max_digits}-digit desegregation (sample):"
				)
				print(list(highest_depth_countries)[:15])

	except Exception as e:
		print(f"Error fetching detailed data: {e}")


scan_highest_digit_nodes()
```
